# 训练模型

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from tensorflow.python.keras.layers.core import Dense, Activation
```

```python
filename = '训练集.csv'
data = pd.read_csv(filename, encoding='gbk')
```

```python
data = data.dropna()  # 去除空集
```

```python
data
```

```python
data.dtypes
```

```python
X = data.iloc[:, 1:15]
X.astype(np.float32)
y = data.iloc[:, 15].astype(int)
```

```python
X
```

```python
y
```

```python
model = Sequential()  # 建立模型
model.add(Dense(input_dim = 14, units = 32))  #
model.add(Activation('relu'))  # 用relu函数作为激活函数，能够大幅提供准确度
model.add(Dense(input_dim = 32, units = 1))
model.add(Activation('sigmoid'))  # 由于是0-1输出，用sigmoid函数作为激活函数
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(X, y, epochs = 1000, batch_size = 10) 
```

```python
e=model.evaluate(X,y,batch_size=10)
w=model.get_weights()
```

```python
def cm_plot(y, yp):  # 混淆矩阵可视化
  
  from sklearn.metrics import confusion_matrix #导入混淆矩阵函数

  cm = confusion_matrix(y, yp) #混淆矩阵
  
  import matplotlib.pyplot as plt #导入作图库
  plt.matshow(cm, cmap=plt.cm.Greens) #画混淆矩阵图，配色风格使用cm.Greens，更多风格请参考官网。
  plt.colorbar() #颜色标签
  
  for x in range(len(cm)): #数据标签
    for y in range(len(cm)):
      plt.annotate(cm[x,y], xy=(x, y), horizontalalignment='center', verticalalignment='center')
  
  plt.ylabel('True label') #坐标轴标签
  plt.xlabel('Predicted label') #坐标轴标签
  return plt
```

# 测试精度

```python
test = pd.read_csv('测试集.csv',encoding='gbk')
```

```python
test = test.dropna()
```

```python
test
```

```python
X = test.iloc[:,1:15]
```

```python
y = test.iloc[:,15]
```

```python
yp=(model.predict(X,batch_size=10) > 0.5).astype("int32").reshape(len(y))
```

```python
cm_plot(y,yp).show()
```

```python
accuracy = float((yp == y).astype(int).sum()) / float(y.size)
```

导入尔康制药15年的财务数据进行预测

```python
y_predict = model.predict(X)
```

```python
yp = y_predict.round(0).astype(int)
yp
```

```python
ls = []
for i in yp:
    i = int(i)
    if i == 1:
        ls.append('违规')
    else:
        ls.append('未违规')
```

```python
erkang['是否违规'] = ls
```
